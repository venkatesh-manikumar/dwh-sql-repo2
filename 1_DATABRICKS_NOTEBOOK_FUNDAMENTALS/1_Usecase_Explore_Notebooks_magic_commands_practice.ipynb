######Task9: Create a python function to convert the given input to upper case

```python
import pyspark.sql.functions as F
def to_upper(df):
    return df.selectExpr("date", "upper(mobile_operating_system) as mobile_operating_system", "percent_of_usage")
rst = to_upper(df)
rst.limit(5).display()
```
